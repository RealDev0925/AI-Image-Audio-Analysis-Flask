```python
# Writes the assets for a given filepath from theme

import re

def update_html_file(filepath):
    try:
        # Read the existing HTML content from the file
        with open(filepath, 'r', encoding='utf-8') as f:
            html_content = f.read()

        print("Successfully read the file.")

        # Regex patterns
        link_pattern = r'href="assets/([^"]+)"'
        link_replacement = 'href="{{ url_for("static", filename="assets/\\1") }}"'
        
        script_pattern = r'src="assets/([^"]+)"'
        script_replacement = 'src="{{ url_for("static", filename="assets/\\1") }}"'

        # Perform the replacements
        updated_content = re.sub(link_pattern, link_replacement, html_content)
        updated_content = re.sub(script_pattern, script_replacement, updated_content)

        print("Successfully performed the replacements.")

        # Write the updated HTML content back to the file
        with open(filepath, 'w', encoding='utf-8') as f:
            f.write(updated_content)
        
        print("Successfully wrote the updated content back to the file.")
    except Exception as e:
        print(f"An error occurred: {e}")

# # Filepath to your HTML template
# filepath = "templates/2.html"

# # Update the HTML file
# update_html_file(filepath)

filepath = "templates/theme_partials/layout/partials/sidebar/_logo.html"
update_html_file(filepath)
```

```python
import re

def extract_partial_path(input_str):
    """
    Extracts the partial path from an HTML comment string.
    
    Args:
        input_str (str): The HTML comment string.

    Returns:
        str: The extracted path if found, otherwise None.
    """
    # Define the regular expression pattern
    pattern = r'<!--layout-partial:(.*?)-->'

    # Use re.search to find the pattern
    match = re.search(pattern, input_str)

    if match:
        # Extract the desired part
        extracted_path = match.group(1)
        return extracted_path
    else:
        return None

# Test the function
input_str = "<!--layout-partial:layout/partials/_content.html-->"
result = extract_partial_path(input_str)

if result:
    print(f"Extracted path: {result}")
else:
    print("Pattern not found")
```

```python
# Get content from demo source file and put it in the new file.

from bs4 import BeautifulSoup

def extract_and_save_div(demo_page_path, div_to_extract_id, save_path, save_filename):
    try:
        # Step 1: Read content from the source HTML file
        with open(demo_page_path, 'r', encoding='utf-8') as file:
            page_content = file.read()

        # Step 2: Use BeautifulSoup to extract the required div
        soup = BeautifulSoup(page_content, 'html.parser')
        div_to_extract = soup.find('div', {'id': div_to_extract_id})

        if div_to_extract:
            extracted_content = str(div_to_extract)
        else:
            print(f"Could not find a div with id={div_to_extract_id}")
            return

        # Step 3: Write the extracted content to the new HTML file
        save_file_path = f"{save_path}/{save_filename}"
        with open(save_file_path, 'w', encoding='utf-8') as file:
            file.write(extracted_content)

        print(f"Extracted content saved to {save_file_path}")
    except FileNotFoundError:
        print(f"File {demo_page_path} not found.")
    except PermissionError:
        print(f"Permission error: Cannot read/write files.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
# extract_and_save_div("keen_v3.0.5/demo1/dist/apps/chat/private.html", "kt_app_content", "templates/partials/pages", "chat.html")


demo_page_path ="keen_v3.0.5/demo1/dist/apps/user-management/users/list.html"
div_to_extract_id = "kt_app_content"
save_path = "templates/partials/pages"
save_filename = "users_list_data_table.html"

extract_and_save_div(demo_page_path, div_to_extract_id, save_path, save_filename)




```
