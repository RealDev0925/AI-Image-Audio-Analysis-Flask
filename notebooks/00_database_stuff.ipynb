```python
from dotenv import load_dotenv
import os
from supabase import create_client, Client
from flask import jsonify


load_dotenv()  # Load environment variables from .env file

# Initialize Supabase
SUPABASE_URL = os.getenv('SUPABASE_URL')
SUPABASE_KEY = os.getenv('SUPABASE_KEY')
supabase = create_client(SUPABASE_URL, SUPABASE_KEY)

def supabase_update (input_dict):
    id = input_dict["id"]
    table = input_dict["table"]
    new_data = input_dict["new_data"]

    data = supabase.table(table).update(new_data).eq("id", id).execute()
    # data = supabase.table("countries").update({"country": "Indonesia", "capital_city": "Jakarta"}).eq("id", 1).execute()

    # print(data)
    # print(data.data)

    data = data.data

    print(data)
    return data



payload = {
    "table": "scenes",
    "id": 1,
    "new_data" : {"scene_text": "7 Sep 2023 at 12:36:58"}
}

update = supabase_update(payload)
# print(update)


```

```python
# get_album_info 


import requests
import json
url = "http://127.0.0.1:5000/supabase/select"
table = "albums"
column = "id"
value = 1

response = requests.get(url, params={"table": table, "column": column, "value": value})

if response.status_code == 200:
    print("Success:", response.json())
else:
    print("Failed:", response.status_code)


response = response.json()

# get pexels response



# make media list

```

```python
full_pexels_response = response[0]["pexels_response"]
print(full_pexels_response)

# dict_keys(['avg_color', 'duration', 'full_res', 'height', 'id', 'image', 'tags', 'url', 'user', 'video_files', 'video_pictures', 'width'])

media_list_prep = []

for item in full_pexels_response:
    # print(item)
    id = item["id"]
    image = item["image"]
    url = item["url"]
    video = item["video_files"][0]["link"]



    media_list_prep.append({"id": id, "image": image, "url": url, "video": video})

print(media_list_prep)




```

```python
# update supabase with media list

import requests
import json

# Define the URL for the request

url = "http://127.0.0.1:5000/supabase/update"

# Define the payload for the request
payload = {
    "table": "albums",
    "id": 1,
    "new_data" : {"media_list": media_list_prep}
}

# Perform the POST request
response = requests.post(url, json=payload)
print(response)

# # Check the response
if response.status_code == 200:
    print("Success:", response.json())
else:
    print("Failed:", response.status_code, response.json())

```
